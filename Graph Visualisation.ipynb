```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
G= nx.Graph()
```

```python
df = pd.read_csv('/Users/shrutisivakumar/Library/CloudStorage/OneDrive-Personal/College Stuff/Sem 1/Projects/Discrete Mathematics/facebook_clean_data/politician_edges.csv')
for node in df['node_1']:
  G.add_node(node)
for source, target in df[['node_1','node_2']].values:
  G.add_edge(source,target,length =100000)
```

```python
nx.draw(G)
plt.show(G)
```
